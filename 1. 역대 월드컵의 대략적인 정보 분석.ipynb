# 1. 역대 월드컵의 대략적인 정보 분석
- 역대 월드컵의 다양한 정보들을 가지고 간단한 분석 활동을 해보겠습니다.

```python
# Pandas 라이브러리 임포트
import pandas as pd
from pandas import Series, DataFrame

# 시각화를 위한 matplotlib 임포트
import plotly.express as px
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
%matplotlib inline
plt.rcParams["font.family"] = "Malgun Gothic"

world_cups = pd.read_csv('data/WorldCups.csv', encoding='CP949')
world_cups

#1등,2등,3등,4등을 많이 한 나라
#개최국에서 입상을 한 비율(ex.개최국 대한민국 입상 대한민국) -> 랭킹 세우기
#경기당 득점수 비율
#관객수
#연도 | 개최지 | 1등  | 2등  | 3등  | 4등 | 총경기 | 득점수 | 참가나라 수 | 총 경기 수 | 관객수
```

```python
#역대 월드컵의 관중 수 출력하기
year_attendance = world_cups[["Year","Attendance"]]
year_attendance
```

```python
#역대 관중 수 시각화해서 나타내기

#그래프를 그리기 위한 판 나타기
plt.rcParams["figure.figsize"] = (6,6)

#yaer 컬럼 값 저장히기
year = year_attendance['Year']

#관객수 컬럼 값 저장하기
attendance = year_attendance['Attendance']

#관객수 그래프 출력
plt.plot(year,attendance,color='black', marker='o')
plt.grid(True)

#관객수라는 값 y 축에 삽입
plt.ylabel('관객수', rotation=360, loc="top")

#연도값 x 축에 삽입
plt.xlabel('연도')

plt.title("역대 관중 수 시각화 내용")
plt.show()
```

```python
#역대 월드컵의 경기당 득점 수
GoalsPerMatch = world_cups[["Year","GoalsScored","MatchesPlayed"]]

#GoalsPerMatch 이라는 컬럼에 골비율값 저장하기
GoalsPerMatch["GoalsPerMatch"] = world_cups["GoalsScored"]/world_cups["MatchesPlayed"]
GoalsPerMatch
```

```python
#2개의 그래프를 출력하기 위한 배열/크기 설정
fig, axes = plt.subplots(2, 1, figsize = [6,12])

#첫 번째 배열의 그래프 설정을 해준다

#모든 경기의 득점 수를 bar그래프로 설정해준다
axes[0].bar(x=GoalsPerMatch['Year'], height=GoalsPerMatch['GoalsScored'], color='grey', label='총 골 득점수')

#지금까지의 경기 참여 팀을 plot팀으로 설정해준다.
axes[0].plot(GoalsPerMatch['Year'],GoalsPerMatch['MatchesPlayed'], marker='o',color='blue', label='모든 경기')

#위에서 설정한 라벨의 위치 지정을 해준다.
axes[0].legend(loc='upper left')
axes[0].grid(True)

#두 번째 배열의 그래프 설정을 해준다

#모든 경기의 득점 수를 barh 그래프로 설정해준다
axes[1].barh(width = GoalsPerMatch['GoalsScored'], y = GoalsPerMatch['Year'],color='green', label='월드컵 골 득점수')

#한 경기당 득점한 골 수를 barh 그래프로 설정해 비율을 확인한다
axes[1].barh(width = GoalsPerMatch['GoalsPerMatch'], y = GoalsPerMatch['Year'],color='red', label='한 경기당 골 득점 수')

#위에서 설정한 라벨의 위치 지정을 해준다.
axes[1].legend(loc='lower right')
axes[1].grid(True)
```

```python
#그래프를 그리기 위한 판 나타기
#모든년도 참가 팀 수 출력하기
plt.rcParams["figure.figsize"] = (6,6)
```

```python
#1등~4등을 몇 번 했는지 나타내기
rankcups = pd.read_csv('data/WorldCups.csv')

#각각의 필요한 컬럼값을 저장
winner = rankcups['Winner'] 
second = rankcups['Runners-Up']
third = rankcups['Third']
fourth = rankcups['Fourth']

#우승 수를 구하기 위한 count 구하기
winner_count = pd.Series(winner.value_counts())
second_count = pd.Series(second.value_counts())
third_count = pd.Series(third.value_counts())
fourth_count = pd.Series(fourth.value_counts())

ranks = pd.DataFrame({
    "Winner" : winner_count,
    "Second" : second_count,
    "Third" : third_count,
    "Fourth" : fourth_count
}) #각각의 컬럼값에 카운트 한 값 지정 후 rank라는 데이터프레임에 지정을 해준다.

ranks = ranks.fillna(0).astype('int64')
#값을 정렬해준다.
ranks = ranks.sort_values(['Winner','Second','Third','Fourth'], ascending=False)
rankss = ranks
count = rankss['Winner'] + rankss['Second'] + rankss['Third'] + rankss['Fourth']
rankss["count"] = count
counts = rankss
counts
```

```python
x = np.array(list(range(0, len(counts))))

# 그래프를 그립니다.
fig, ax = plt.subplots()

# x 위치에, 항목 이름으로 ranks.index(국가명)을 붙입니다.
plt.xticks(x, ranks.index, rotation=80)
plt.tight_layout()

# 4개의 막대를 차례대로 그립니다. 한 값에 여러 그래프를 넣기 위해 x좌표를 지정해준다.
ax.bar(x - 0.3, ranks['Winner'],     color = 'gold',   width = 0.2, label = '1등')
ax.bar(x - 0.1, ranks['Second'], color = 'silver', width = 0.2, label = '2등')
ax.bar(x + 0.1, ranks['Third'],      color = 'brown',  width = 0.2, label = '3등')
ax.bar(x + 0.3, ranks['Fourth'],     color = 'black',  width = 0.2, label = '4등')

#x,y 축의 이름을 설정해준다.
plt.xlabel('나라')
plt.ylabel('우승횟수',rotation=360, loc="top")
plt.legend(loc='upper right')
```

```python
#X축 인덱스 값 지정
plt.xticks(x, counts.index, rotation=80)

plt.title("총 입상 횟수")

#count 컬럼 값 출력
plt.bar(height = counts['count'], x = counts.index,color='red', label='총 입상 수')
plt.legend(loc='upper right')

#입상 수 라는 값 y 축에 삽입
plt.ylabel('입상', rotation=360, loc="top")

#나라라는 값 x 축에 삽입
plt.xlabel('나라')
```

```python
#개최국에서 자신의 나라가 1등한 경우와 그나라의 관객수와 해당 올림픽 첨전 나라 수
#[Year,Country,Winner,Attendance]
ww = rankcups[['Year','Country','Winner','Attendance','QualifiedTeams']]
www = ww['Country'] == ww['Winner']
data = ww[www]
data
```

```python
#그래프를 그려주기 위한 크기 지정
plt.rcParams["figure.figsize"] = (6,6)
#위 내용의 해당되는 나라를 x축으로 지정해준다(자신의 개최국에서 우승)
xindex = data['Winner'] = ['우루과이','이탈리아','잉글랜드','독일','아르헨티나','프랑스']

#x축인 우승한 나라 우승한 나라의 연도를 y축으로 지정해준다.
plt.plot(xindex, data['Year'], marker='o',color='blue', label='우승연도')
plt.legend(loc='lower right')

#제목,xlabel 지정
plt.title("자신의 홈 개최지에서 우승한 나라")
plt.xlabel("나라명")
plt.ylabel('연도', rotation=360, loc="top")

plt.grid(True)
```

```python
#우승연도의 관객 수

plt.rcParams["figure.figsize"] = (6,6)

#위 내용의 해당되는 나라를 x축으로 지정해준다(자신의 개최국에서 우승)
xindex = data['Winner'] = ['우루과이','이탈리아','잉글랜드','독일','아르헨티나','프랑스']

#관객수 값을 plot그래프로 표현해준다.
plt.plot(xindex, data['Attendance'], marker='o',color='red', label='관객 수')
plt.legend(loc='lower right')

#제목,xlabel 지정
plt.title("자신의 나라에서 우승할 때 관객 수")
plt.xlabel("나라명")
plt.ylabel('관객수', rotation=360, loc="top")

plt.grid(True)
```

```python
world_match = pd.read_csv('data/WorldCupMatches.csv', encoding='CP949')
```

```python
#groupby를 이용해 팀 이름을 기준으로 골 합을 구한다
home = world_match.groupby(['Home Team Name'])["Home Team Goals"].sum()
away = world_match.groupby(['Away Team Name'])["Away Team Goals"].sum()

# home과 away를 열을 기준으로 합치고 결측치를 모두 0으로 초기화해준다.
goal_per_country = pd.concat([home, away], axis=1, sort=True).fillna(0)

# Home Team Goals와 Away Team Goals를 이용해서 총 골 수 구하기
goal_per_country['Goals'] = goal_per_country['Home Team Goals'] + goal_per_country['Away Team Goals']

# Goals을 기준을 내림차순 정렬
goal_per_country = goal_per_country["Goals"].sort_values(ascending = False)

# 값을 정수형으로 변경
goal_per_country = goal_per_country.astype(int)


#상위 10개만 출력 설정해준 후 x축과 y축 값을 설정해준다.
goal_per_country = goal_per_country[:10]
country = goal_per_country.index
goals = goal_per_country.values
fig, ax = plt.subplots()

# 크기 조절을 위해 width를 원래의 50%로 조정해준다.
ax.bar(country, goals, width = 0.5)

# 반시계 방향으로 30도 회전
plt.xticks(country, rotation=30)

# 글자가 넘쳐서 잘리는 현상을 방지해준다.
plt.tight_layout()

plt.grid(True)
```

```python
#2014 연도에 해당되는 내용만 저장해준다
world_match =world_match[world_match['Year']==2014]

#groupby를 이용해 팀 이름을 기준으로 골 합을 구한다
home_team_goal = world_match.groupby(['Home Team Name'])['Home Team Goals'].sum()
Away_team_goal = world_match.groupby(['Away Team Name'])['Away Team Goals'].sum()

# home과 away를 열을 기준으로 합치고 결측치를 모두 0으로 초기화해준다.
team_goal_2014 = pd.concat([home_team_goal,Away_team_goal],axis = 1).fillna(0)

# Home Team Goals와 Away Team Goals를 이용해서 2014년 총 골 수 구하기
team_goal_2014['goals'] = team_goal_2014['Home Team Goals'] + team_goal_2014['Away Team Goals']
team_goal_2014 = team_goal_2014.drop(['Home Team Goals', 'Away Team Goals'], axis=1)

# Goals을 기준을 내림차순 정렬
team_goal_2014 = team_goal_2014['goals'].sort_values(ascending=False)

# 값을 정수형으로 변경
team_goal_2014.astype('int')
team_goal_2014
```

```python
#2014년 나라별 골득점 데이터를 출력해준다
team_goal_2014.plot(x=team_goal_2014.index, y=team_goal_2014.values, kind="bar", figsize=(12, 12), fontsize=14)
```
